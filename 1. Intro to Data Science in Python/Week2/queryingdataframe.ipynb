querying DataFrames - A Boolean mask is an array which can be of one dimension like a series, or two dimensions like a data frame, where each of the values in the array are either true or false. This array is essentially overlaid on top of the data structure that we're querying. And any cell aligned with the true value will be admitted into our final result, and any cell aligned with a false value will not.

```python
import pandas as pd
df=pd.read_csv('resources/Admission_Predict.csv', index_col=0)
df.columns=[x.lower().strip() for x in df.columns]
df.head()
```

```python
admit_mask=df['chance of admit'] > 0.7
admit_mask
```

```python
df.where(admit_mask).head() 
```

```python
df.where(admit_mask).dropna().head()
```
