```python
from matplotlib.backends.backend_agg import FigureCanvasAgg # a backend
from matplotlib.figure import Figure # bring in the fig object

fig = Figure() # a new empty figure

canvas = FigureCanvasAgg(fig) # associate the fig w the canvas backend

ax = fig.add_subplot(111) # adding a subplot directly & 111 means that we want 1 plot. the return value for the subplot is the Axes object, which contains methods for plotting

ax.plot(3, 2, '.')

canvas.print_png('test.png')
```
